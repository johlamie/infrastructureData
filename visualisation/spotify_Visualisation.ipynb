## Récupération des données stockées dans le JSON

```python
with open("../src/collect/data/2020-12-12.json", "r") as read_file:
    track_results = json.load(read_file)
```

## Traitement 

```python
artist_name = []
track_name = []
popularity = []
track_id = []
release_date= []

for i in range(0,10):
    track_results = sp.search(q='year:2020', type='track', limit=50,offset=i)
    for i, t in enumerate(track_results['tracks']['items']):
        artist_name.append(t['artists'][0]['name'])
        track_name.append(t['name'])
        track_id.append(t['id'])
        popularity.append(t['popularity'])
        release_date.append(t['album']['release_date'])
        
    
```

```python
import pandas as pd

df_tracks = pd.DataFrame({'artist_name':artist_name,'track_name':track_name,'track_id':track_id,'popularity':popularity})
```
